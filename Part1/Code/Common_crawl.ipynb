# Script for Commoncrawl

Here we used a warc file from CC News.


We used the url lib request and then beautiful soup to find the content.`

The keyword that matches with the content of the url is stord into a csv file.

Keyword for collecting Wrestling Keyword

```python
from requests.exceptions import ConnectionError
import time
import csv
import warc
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd

f = warc.open("demowarc.warc")
c=0;
mycolumns = ['links']
df = pd.DataFrame(columns=mycolumns)
for record in f:
    try:
        if(record.header['WARC-Type']=='response'):
               #print(record.header['WARC-TARGET-URI'])
                
            #We searched in the content to find any matching key words that will match with the content of the url's
            link1=record.header['WARC-TARGET-URI']
            the_word = 'wrestling'
            r = requests.get(link1, allow_redirects=False)
            soup = BeautifulSoup(r.content, 'lxml')
            words = soup.find(text=lambda text: text and the_word in text)
            print(link1)
            print(c)
            if not words:
                continue
            else:
                df.loc[len(df)] = link1
                c=c+1
                if(c==100):
                    break
        df.to_csv("wrestling.csv", sep='\t',index = False)
        #                 with open('football4.csv', 'a') as writeFile:
        # #                     writer = csv.writer(writeFile)
        # #                     print(link1)
        # #                     writer.writerows(link1)
        # #                     writeFile.close()
        # #                     c=c+1
        # #                     print("count: {}" .format(c

        #                     if(c==100):
        #                         break

    except Exception as e:
        
        #time.sleep(2)
        print(e)
        continue
```
